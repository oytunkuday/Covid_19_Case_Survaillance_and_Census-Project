###Hypothesis Testing

Null hypothesis=In our dataset, among all age groups, being in the 20-29 age group and passing from corona is 2%. 

Alternative hypothesis=In our dataset,  among all age groups, being in the 20-29 age group and passing from corona is higher than 2% 

```
# Sample size
n = 5000

# Number of samples
num_samples = 2000

death_rates = []

# For each sample
for i in range(num_samples):
  sample = np.random.choice([0, 1], p=[0.98, 0.02], size=n)
  death_rate_in_sample = sample.sum() / len(sample)
  death_rates.append(death_rate_in_sample)

death_rates=np.array(death_rates)
```

```
fig, ax = plt.subplots(figsize=(30, 5))
sns.countplot(death_rates, ax=ax)
```

```
p_value = len(death_rates[death_rates >= 0.02]) / len(death_rates)
print("p-value: ", p_value)
```

As one can understand from the p value, we could not reject the null hypothesis.

This means in our dataset, among all age groups, being in the 20-29 age group and passing from corona is 2%. 

Null hypothesis=In our dataset, among all age groups, being in the 60-69 age group and passing from corona is 10%. 

Alternative hypothesis=In our dataset,  among all age groups, being in the 60-69 age group and passing from corona is higher than 10%

```
# Preparing Data for Modelling
```

```
categorical_columns = ['current_status', 'sex', 'age_group', 'Race and ethnicity (combined)', 'hosp_yn', 'icu_yn', 'medcond_yn', 'death_yn']
```

```
df = data.copy()
df = df.dropna()
df = df[~df['current_status'].isin(['Missing', 'Unknown'])]
df = df[~df['hosp_yn'].isin(['Missing', 'Unknown'])]
df = df[~df['icu_yn'].isin(['Missing', 'Unknown'])]
df = df[~df['medcond_yn'].isin(['Missing', 'Unknown'])]
df = df[~df['sex'].isin(['Missing', 'Unknown'])]
df = df[~df['death_yn'].isin(['Missing', 'Unknown'])]
```

```
from sklearn.preprocessing import LabelEncoder

lb_make = LabelEncoder()

for column in categorical_columns:
    df[column] = lb_make.fit_transform(df[column].astype(str))
```

```
#Selecting correlated columns for modelling:
plt.subplots(figsize=(10,10))
sns.heatmap(df.corr(),annot=True,annot_kws={'size':9})
```

From the above heatmap it can be observed that there are correlation between columns age_group,icu_yn,death_yn,hosp_yn,Race and etncity(combined) and medcond_yn therefore we inculed those columns for modelling.

```
X =  df[['current_status','hosp_yn','icu_yn','medcond_yn','sex']]
y = df['death_yn']
from sklearn.model_selection import train_test_split

X_train, X_remaining, y_train, y_remaining = train_test_split(X, y, test_size=0.30, random_state=0,stratify=y)
X_test, X_val, y_test, y_val = train_test_split(X_remaining, y_remaining, test_size=0.50, random_state=0, stratify=y_remaining)
```

We allocated two thirds of our data for training the model and the rest remained for testing.

Fistly we use random forest machine learning algorithm. We use balanced class weight.

# RandomForest

```
from sklearn.ensemble import RandomForestClassifier
rf_c=RandomForestClassifier(class_weight='balanced', n_estimators=100)
pred_rf=rf_c.fit(X_train, y_train)
#Predicting values
Y_pred_rf = pred_rf.predict(X_test)
```

```
from sklearn.metrics import auc, precision_recall_curve

n_estimators = [50,100,200,300,500,750]
max_features = [3,5,7,10,15,20,25]
auprc_estimators = []
auprc_features = []

for estimation in n_estimators:
  newOne= RandomForestClassifier(n_estimators=estimation, random_state=0, class_weight='balanced')
  newOne.fit(X_train,y_train)
  predictions_newone=newOne.predict_proba(X_val)[:,1]
  precision, recall, thresholds = precision_recall_curve(y_val,predictions_newone)
  auprc_estimators.append(auc(recall,precision))

for feature in max_features:
  newOne=RandomForestClassifier(max_features=feature, random_state=0, class_weight='balanced')
  newOne.fit(X_train,y_train)
  predictions_newone=newOne.predict_proba(X_val)[:, 1]
  precision,recall,thresholds=precision_recall_curve(y_val, predictions_newone)
  auprc_features.append(auc(recall,precision))
```

As can be seen below, nothing has changed at all with Hyper Parameter Tuning.
